## 판다스

```python
# 데이터를 조작/분석하기위한 라이브러리임.
# 다양한 데이터 분석 함수를 제공힘.
# 행과 열로 이루어진 객체를 만들고 다룰 수 있음 - ? 3차원의 데이터는 다룰 수 없나?
# 판다스의 데이터 구조는 시리즈와 데이터 프레임이 있음.
```

### 시리즈

```python
# 시리즈는 1차원 배열의 형태로 한가지 형의 데이터로 구성되며, 인덱스는 특별히 지정하지 않으면 숫자 0부터 시작함.
# 보통 인덱스는 문자로 지정함.
```

### 데이터 프레임

```python
# 2차원 배열의 형태로 이루어진 데이터 구조임(엑셀과 비슷함).
# 인덱스와 컬럼이라는 두 가지 기준에 의해 표와 같이 데이터가 저장됨.
# 시리즈가 합쳐진 형태임.
# 데이터 프레임은 이중리스트나 딕셔러니로 만들 수 있음.
# 딕셔너리로 데이터 프레임을 만들면 키(key value)가 컬럼 명으로 지정되어 만들어짐. 
```

```python
# df.columns = 컬럼명 확인
# df.columns = ['컬럼명 1','컬럼명 2','컬럼명 3','컬럼명 4'] 
```

```python
# 데이터 파일: CSV
# 하나의 파일로 관리 가능한 크기의 데이터라면 일반적으로 csv포맷을 활용함.
# 데이터가 콤마로 분리되어 저장.
# 한글이 포함된 경우 인코딩 방식을 utf8로 해야지 한글이 깨지지 않음.
# 데이터 파일의 이름을 적어줄 때 같은 폴더 안에 위치 하지 않는다면, 경로명까지 정확히 적어야함.
# CSV파일을 읽어서 데이터프레임 객체로 만들어서 반환함. (보통 df라는 변수명을 사용함)  
```

```python
# 분석 절차
# 1. 데이터 정보 확인 = 행과 열의 크기 컬럼명, 컬럼을 구성하는 값의 자료형 등을 요약하여 정보를 출력함.
# ex) 데이터 프레임명.info()
# 2. 칼럼별 통계량 = 숫자형 변수에 대한 기초 통계량을 출력함.
# ex) 데이터프레임명.describe()
# 3. 데이터
```

```python
# 미션
# sales, user 두 가지 다 로드할 것 
import pandas as pd
import numpy as np



sales_df = pd.read_csv('sales_data.csv')
user_df = pd.read_csv('user_data.csv')
pd.options.display.float_format = '{:.2f}'.format
```

```python
# 컬럼명을 한글로 변경
# '고객ID', '구매일자', '제품분류', '구매금액', '구매채널', '할인율'
# '고객ID', '고객명', '가입지역'
#sales_df = sales_df.rename(columns = {'customer_id':'고객 ID', 'order_date':'주문일자', 
#						   'product_category':'제품분류', 'amount':'구매금액','channel':'구매채널',
#                           'discount_rate':'할인율'})
#user_df = user_df.rename(columns= {'customer_id':'고객ID', 'customer_name': '고객명',
#						 'region': '지역'},)		
sales_df.columns = ['고객ID', '구매일자', '제품분류', '구매금액', '구매채널', '할인율']
user_df.columns = ['고객ID', '고객명', '가입지역']
```

```python
sales_df
```

```python
# 구매금액이 10000원 이상인 데이터만 필터링
sales_df[sales_df['구매금액'] > 10000]
```

```python
# '온라인' 채널에서 발생한 구매 중 5000원 이상인 것만 필터링
sales_df.loc[(sales_df['구매금액'] > 5000) & (sales_df['구매채널'] == '온라인')]
```

```python
# '가전', '주방용품' 에 속하는 데이터만 필터링
#sales_df.loc[(sales_df['제품분류'] == '가전') | (sales_df['제품분류'] == '주방용품')] = 가짓수 적을때
sales_df[sales_df['제품분류'].isin(['가전', '주방용품'])] # 가짓수가 많을때
```

```python
# 각 열별 결측치 개수 확인
sales_df.isnull().sum()
```

```python
# 결측치가 있는 행 전체 삭제
sales_df.dropna()
```

```python
# 특정 값으로 결측치 채우기
# sales_df = sales_df.fillna(method='ffill')
sales_df['할인율'] = sales_df['할인율'].fillna(0)
```

```python
sales_df
```

```python
# 기존 열을 이용해 새로운 열 생성 
# 할인률이 적용된 df['실구매금액'] 새롭게 생성
sales_df['실구매금액'] = sales_df['구매금액'] * (1 - sales_df['할인율'])
sales_df
```

```python
# 고객 등급을 'VIP'와 '일반'으로 나누기(구매금액이 10000 이상이면 VIP)
sales_df['고객등급'] = sales_df['구매금액'].apply(lambda x: 'VIP' if x >= 10000 else '일반')
```

```python
# 제품분류별 구매금액의 평균 구하기
print("8 & 9. 실구매금액 및 고객등급 생성:")
print(sales_df[['구매금액', '할인율', '실구매금액', '고객등급']].head())
print("-" * 30)
```

```python
# 구매채널별로 그룹화하여, 구매금액의 총합과 평균 동시 구하기
channel_stats = sales_df.groupby('구매채널')['구매금액'].agg(['sum', 'mean'])
print("\n11. 구매채널별 구매금액 총합 및 평균:")
print(channel_stats)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
 
# --- 데이터 생성---
data = {
    '월': pd.to_datetime(['2025-01-01', '2025-02-01', '2025-03-01', '2025-04-01', '2025-05-01', '2025-06-01']),
    '방문자수': [1500, 1800, 1700, 2100, 2500, 2300]
}
df_line = pd.DataFrame(data)
 
plt.figure(figsize=(10, 6)) # figsize 설정
 
# 1. 플롯 그리기
plt.plot(
    df_line['월'],
    df_line['방문자수'],
    marker='o',
    linestyle='-',
    color='dodgerblue'
)
 
# 2. 그래프 꾸미기 (ax.set_ 대신 plt. 사용)
plt.title('2025년 상반기 월별 웹사이트 방문자 수', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('방문자 수 (명)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
 
data = {
    '부서': ['영업1팀', '영업2팀', '마케팅팀', '기술지원팀'],
    '실적(억원)': [12, 15, 8, 5]
}
df_bar = pd.DataFrame(data)
 
plt.figure(figsize=(10, 6))
 
bars = plt.bar(
    df_bar['부서'],
    df_bar['실적(억원)'],
    color=['#FF5733', '#33FF57', '#3357FF', '#FF33A1']
)
 
plt.title('2023년 3분기 부서별 판매 실적', fontsize=16)
plt.xlabel('부서', fontsize=12)
plt.ylabel('실적 (억원)', fontsize=12)
 
# 각 바 위에 실제 값 표시 (plt.text 사용)
for bar in bars:
    yval = bar.get_height()
    # plt.text(x좌표, y좌표, 내용, 정렬방식)
    plt.text(
        bar.get_x() + bar.get_width()/2.0,
        yval,
        f'{yval}억',
        va='bottom',
        ha='center'
    )
 
plt.show()
```

```python
# 추천은 아님
import matplotlib.pyplot as plt
import numpy as np
 
data = np.random.rand(8, 10)
rows = [f'Row{i+1}' for i in range(data.shape[0])]
cols = [f'Col{i+1}' for i in range(data.shape[1])]
 
plt.figure(figsize=(10, 7))
im = plt.imshow(data, cmap='coolwarm')
plt.colorbar(im)
plt.xticks(np.arange(data.shape[1]), labels=cols)
plt.yticks(np.arange(data.shape[0]), labels=rows)
plt.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
plt.setp(plt.gca().get_xticklabels(), rotation=-30, ha="right", rotation_mode="anchor")
 
# 각 셀에 텍스트 추가 (Annotation)
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        # plt.text(x, y, s) 순서입니다.
        plt.text(j, i, f'{data[i, j]:.2f}',
                 ha="center", va="center", color="w")
 
plt.title("Heatmap using Matplotlib", pad=20)
plt.tight_layout()
plt.show()
```
