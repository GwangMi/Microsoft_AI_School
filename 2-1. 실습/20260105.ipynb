#### range라는 함수가 있음

```python
for i in range(1, 6, 2):
    print(f'{i}hello')
```

#### range괄호안의 start, stop, step 인데 start는 시작점, stop는 끝점, Step는 건너뛰기 라고 생각하면됨

```python
# 1~10까지 합의 결과를 구하는 반복문

sum = 0
for i in range(1,11):
    sum =  (sum + i)

print(sum)

    
```

```python
# 업다운 게임(이전에 내가 작성한 코드)
import random

true_value = random.randint(1,101)
input_value = 999999

print("숫자를 맞춰보세요(1~100)")
while true_value != input_value:
    input_value = int(input())

    if input_value > true_value: #사용자의 입력값이 ture_value보다 클 때
        print("숫자가 너무 큽니다.")
    elif input_value < true_value: #사용자의 입력값이 ture_value보다 작을 때
        print("숫자가 너무 작습니다.")
    else: 
        break
print(f"정답입니다. 입력한 숫자는 {true_value}")
```

```python
import random
answer = random.randint(1,101)

while True:
    guess = int(input("1~100사이의 정수를 입력하세요."))
    if answer == guess:
        print('정답입니다.')
        break
    elif answer < guess:
        print('숫자가 너무 큽니다')
    else:
        print('숫자가 너무 작습니다.')
print('게임종료')
```

### 오늘의 미션

```python
#사용자로 부터 n개의 별의 갯수를 입력받은 뒤
# 별을 출력하되 별의 갯수만큼만 출력하다가 별이 다 소진되면 멈추는 코드 작성
# 트리의 최대높이는 최대 10층으로 한정한다.
star = int(input('별의 갯수를 입력해주세요'))
used_star = 0

for i in range(1,11):
    # print('*' * i)
    star_remain = star - used_star
    if star_remain > i:
        print("*" * i)
        used_star += i
    else:
        print("*" * star_remain)
        break
```
