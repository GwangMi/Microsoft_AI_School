```python
numbers = [4,4,4,3,3]
result = []
compare_num = []
result.append(compare_num)

for i in numbers:
    if compare_num != i:
        compare_num = i
        result.append(compare_num)
    

print(result)
```

```python
# [미션] 전화번호 검색프로그램을 다음과 같이 만들어 볼 것.
# 추가, 검색 기능을 구현할 것.
# 1. 메뉴선택 (1.추가, 2.검색, 0.종료)

addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234'
}


while True:
    x = int(input('해당하는 메뉴를 눌러주세요 1. 추가, 2. 검색, 0 종료'))
    if x == 1:
        y = input('이름을 입력하여주세요.')
        z = input('전화번호를 입력하여 주세요')
        addr[y] = z
        print('추가되었습니다')
        
    elif x == 2:
        print("친구들 목록:", list(addr.keys()))
        a = input('이름을 입력하여 주세요')
        if a in addr:
            print(a,addr[a])
        else:
            print('그런 사람은 없습니다.')
        pass
    elif x == 0:
        print('프로그램이 종료되었습니다.')
        break
    else:
        print('잘못된 번호입니다.')
```

```python
# 모스 부호 해석 프로그램 만들기
# 입출력 예
# letter  result
# ".... . .-.. .-.. ---"  "hello"
# ".--. -.-- - .... --- -."   "python"

# 모스부호는 다음과 같습니다.
morse = {
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f',
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l',
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r',
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x',
    '-.--':'y','--..':'z'
}

code = input("모스 부호 입력")

letters = code.split(" ")
result = ""

for i in letters:
    result += morse[i]

print("해석 결과", result)
```

```python
def num_max(a, b, c):
    return max(a, b, c)

num = num_max(10, 12, 7)
num
```

```python
def num_max(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val

num = num_max(10, 12, 7)
num
```

```python
min_num = [-1,7,12,102,111,-52]

def num_min(array):
    min_val = array[0]
    for i in array:
        if min_val > i:
            min_val = i
    return min_val

num = num_min(min_num)
num
```

```python
# 윤년을 확인하는 함수 만들기
# 4로 나누어 떨어지는 해는 윤년임.
# but 100으로 나누어 떨어지는 해는 윤년이 아님.
# but 400나누어 떨어지는 해는 윤년임.

def leaf_years(year):
    return(year % 4 == 0) and (year % 100 != 100) or (year % 400 == 0)

leaf_years(2025)

```

```python
def add_mul( choice , *args):
    print(choice, args)
    if choice == 'add':
        result = 0
        for i in args:
            result += i
    elif choice == 'mul':
        result = 1
        for i in args:
            result *= i
    else:
         print('연산자 종류 확인 요망')
         result = None
    return result

```

#### 람다함수

```python
# 일반적인 함수는 함수의 이름과 매개변수, 리턴값을 사용하여 작성함.
# 하지만 람다함수는 이름이 없음
# 장점으로는 코의 간결함과 메모리가 절약됨.
```

```python
mylist = [1,2,3,4,2,5,6,3,4]
result = map(lambda x : x*2, mylist)
list(result)
```

```python
numbers = [1,2,3,4,5,6]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
numbers = [1,2,3,4,5,6]
squared = list(filter(lambda x: x % 2 ==0, numbers))
print(squared)
```

```python
point = [(1, 2), (3, 4), (5, 1)]

result = sorted(point, key = lambda x: x[1])

print(result)
```
