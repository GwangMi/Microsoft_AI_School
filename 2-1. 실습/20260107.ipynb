```python
# 등수구하기 - 평균점수
score1 = [[80, 70], [90, 50], [40, 70], [50, 80]]
print(rank(score1)) # [1, 2, 4, 3]
 
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
print(rank(score2)) # [4, 4, 6, 2, 2, 1, 7]
 
def rank(score):
    avg = [(i + j)/2 for i , j in score]
    avg_sort = sorted(avg, reverse = True)
    result = [avg_sort.index(i)+1 for i in avg]
    return result
```

### 클래스

```python
# 클래스란?
# 객체를 표현하기 위한 문법임, 여기서의 객체는 체크박스나, 스크롤바, 터틀객체 등의 특정한 모양이나 개념이 존재하는 것을 말함
# 현실세계 사물의 속성과 동작들을 코드로 구현하는 것임.
# 관련된 속성과 동작을 하나의 범주로 묶음.
# 사물의 속성은 변수로, 동작은 함수(메소드)로 표현함.
#   - 속성은 매개변수를 받아 사용하기 위한 값을 정하
#   - 메소드는 만들어진 속성을 이용해 그에 따른 동작을 실행함.
# 클래스를 구성하는 변수와 함수를 멤버라고 하며 클래스에 속한 함수를 메소드라고 한다.
```

```python
# 클래스는 똑같은 것을 께속 만들어 낼 수 있는 틀, 설계도 등으로 정의하기도 함.
# 클래스에서 찍어 낸 결과물을 객체라고 하며, 객체마다 고유한 성격을 가짐.
# 클래스의 인스턴스(instance)라고 표현하기도 함. 
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def walk(self):
        print(f'{self.name}이 걸어요')
        
```

```python
person1 = Person("박지민", 31)
person1
```

```python
# 상속
class NewPerson(Person):
    def fly(self): #기능확장
        print(f'{self.name}이 날아요')
    def walk(self): # 매서드 오버라이딩
        print(f'{self.name}이 빨리 뛰어요')
```

```python
person3 = NewPerson("민윤기", 33)
person3.fly()
```

```python
class NewNewPerson(Person):
    def __init__(self, name, age, hobby):
        super().__init__(name, age)
        self.hobby = hobby
    
    def fly(self): #기능확장
        print(f'{self.name}이 날아요')
    
    def walk(self): # 매서드 오버라이딩
        print(f'{self.name}이 빨리 뛰어요')

    def introduce(self):
        print(f'{self.name}의 취미는 {self.hobby}입니다.')
```

```python
person4 = NewNewPerson("김태형", 33 "노래부르기")
```
