```python
# coding: utf-8
import os, sys
print(os.getcwd())
current_dir = os.path.dirname(os.getcwd())
print(current_dir)
os.chdir(current_dir)

import numpy as np
import matplotlib.pyplot as plt
from ch06.simple_convnet import simpleconvnet

def filter_show(filter, nx=8, margin=3, saale=10):
    """""
    c.f. https://gist.github.com/aidiary/07d530d5e08011832bb12#file-draw_weight-py
    """""

    FN,C,FH,FW = filter.shape
    ny = int(np.ceil(FN/nx))

    fig = plt.figure()
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

    for i in range(FN):
        ax = fig.add_subplot(ny, nx, i+1, xticks=[], yticks=[])
        ax.imshow(filter[i,0], cmap=plt.cm.gray_r, interpolation='nearest')
    plt.show()

network = SimpleConvNet()


network.load_params("ch06/params.pkl")
filter_show(network.params['W1'])
```
