### 모델 만들기

```python
# plot tree
# 특성 중요도 시각화
# 레포트
```

```python
# TODO: 데이터 분리하기
X = rocket.drop('Launched?', axis=1)
y = rocket['Launched?']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

rocket_dt = DecisionTreeClassifier(random_state=42)

rocket_dt.fit(X_train, y_train)
```

```python
print(rocket_dt.score(X_train, y_train))
print(rocket_dt.score(X_test, y_test))
```

```python
pred = rocket_dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(18, 7))
plot_tree(rocket_dt, filled=True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.barplot(y = rocket_dt.feature_names_in_, x= rocket_dt.feature_importances_)
```

```python
# TODO: 모델 만들고 성능 평가하기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# 데이터 분리하기
X = rocket.drop('Launched?', axis=1)
y = rocket['Launched?']

# 학습모델 및 테스트 모델 분류
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)


```

```python
# 결정트리모델분류
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
dt_y_pred = dt.predict(X_test)
print(classification_report(y_test,dt_y_pred))
```

```python
# 랜덤포레스트
rt = RandomForestClassifier(random_state=42)
rt.fit(X_train,y_train)
print(rt.score(X_train, y_train))
print(rt.score(X_test, y_test))
rt_y_pred = rt.predict(X_test)
print(classification_report(y_test,rt_y_pred))
```

```python
# 로지스틱 회귀
lr = LogisticRegression(random_state=42)
lr.fit(X_train,y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
lr_y_pred = lr.predict(X_test)
print(classification_report(y_test,lr_y_pred))
```

```python
# KNN 알고리즘
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
knn_y_pred = knn.predict(X_test)
print(classification_report(y_test,knn_y_pred))
```

### GridSearchCV

```python
# GridsearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state = 42)

parameters = {'max_depth':[3,4,5],
              'min_samples_split':[2,4]}

gscv = GridSearchCV(rocket_dt, param_grid=parameters, cv=skf, refit=True)

gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
best_model = gscv.best_estimator_
pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(best_model.score(X_train,y_train))
print(best_model.score(X_test,y_test))
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(18, 7))
plot_tree(best_model, filled=True)
plt.tight_layout()
plt.show()
```
